```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
# Take last 10 fake news rows for manual testing
data_fake_manual_testing = data_fake.tail(10)
data_fake = data_fake.iloc[:-10]

# Take last 10 true news rows for manual testing
data_true_manual_testing = data_true.tail(10)
data_true = data_true.iloc[:-10]
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing['class'] = 0
data_true_manual_testing['class'] = 1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
import re, string

def wordopt(text):
    text = text.lower()                                    # lowercase
    text = re.sub(r'\[.*?\]', '', text)                    # remove things in brackets
    text = re.sub(r'https?://\S+|www\.\S+', '', text)      # remove urls
    text = re.sub(r'<.*?>+', '', text)                     # remove html tags
    text = re.sub(r'\n', ' ', text)                        # remove line breaks
    text = re.sub(r'\s+', ' ', text).strip()               # remove extra spaces
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=42, stratify = y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def predict_news(news):
    news = wordopt(news)
    vectorized = vectorization.transform([news])
    prediction = LR.predict(vectorized)[0]
    prob = LR.predict_proba(vectorized)[0]
    return f"{'Fake' if prediction == 0 else 'Real'} (Confidence: {max(prob)*100:.2f}%)"

# Try with your manual test set
print(predict_news("Donald Trump wins the election again in 2024!"))
print(predict_news("NASA announces new evidence of water on Mars."))
```
